```python
df = pd.read_csv('1quarter_df.csv')
df.head()
```

```python
df.dropna(inplace = True)
```

```python
df.isnull().sum()
```

```python
#df.drop([df['openDt'] == " "], axis = 0)
df = df[df['openDt'] != " "]
```

```python
df[df['movieCd']=='불빛 아래서']
```

```python
# 흥행 정도를 (1 2 3)으로 나누기 : 관객 수 기준 최소가 0명 최대가 1000000명 3300000 660000이 기준이 되는 거죠
# openDt와  load_dt의 차이를 하나의 칼럼으로 넣기 => 상영 기간
# 1인 것들 2인 것들 3인 것들의 평균상영기을 구하기 
# 날짜를 datetime 형식으로 변환

df['openDt'] = pd.to_datetime(df['openDt'], format='%Y-%m-%d')
df['load_dt'] = pd.to_datetime(df['load_dt'], format='%Y%m%d')
print(df['openDt'])
#print(df['load_dt'])
```

```python
df = df[df['openDt'] > '2016-11-01']
```

```python
df.columns
```

```python
df.dtypes
```

```python
# 상영 기간 추가
df['screening_period'] = (df['load_dt'] - df['openDt']).dt.days
df.head()
```

```python
#df['screening_period'] = df[df['screening_period'] > 0]
df = df[df['screening_period'] > 0]
df = df[df['screening_period'] < 80 ]
```

```python
# quantile을 사용하여 흥행 정도 나누기
q1 = df['audiCnt'].quantile(0.33)
q2 = df['audiCnt'].quantile(0.66)

def classify_audience(audiCnt):
    if audiCnt <= q1:
        return 1
    elif audiCnt <= q2:
        return 2
    else:
        return 3

df['popularity'] = df['audiCnt'].apply(classify_audience)

# 흥행 정도별 관객 수 평균 구하기
avg_audience_by_popularity = df.groupby('popularity')['audiCnt'].mean()

print(df)
#print("\n흥행 정도별 관객 수 평균:")
#print(avg_audience_by_popularity)
```

```python
df.head()
```

```python
df['popularity'].value_counts()
```

```python
df_pop1 = df[df['popularity'] == 1]
df_pop2 = df[df['popularity'] == 2]
df_pop3 = df[df['popularity'] == 3]
```

```python
df_pop1g = df_pop1.groupby(['movieNm'])['screening_period'].max().to_frame().reset_index()
df_pop2g = df_pop2.groupby(['movieNm'])['screening_period'].max().to_frame().reset_index()
df_pop3g = df_pop3.groupby(['movieNm'])['screening_period'].max().to_frame().reset_index()
```

```python
df_pop1g
```

```python
df2 = df.groupby(['popularity','movieNm'])['screening_period'].max().to_frame()
avg_screening_period = df2.groupby(level=0)['screening_period'].mean().to_frame()
avg_screening_period
#df2
```

```python
avg_screening_period.reset_index(inplace = True)
```

```python
avg_screening_period
```
